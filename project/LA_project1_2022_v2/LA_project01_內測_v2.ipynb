## 題目2：請用以下2種方法得到矩陣A 並printA的size 大小.(30pt)

### i.e. 宣告一個形態為ndarray的一維陣列Element：[1 0 0 0 0 0 3 0 0 0 0 0 5 0 0]
###       方法1.請以for迴圈將zero_A內的元素依次替換成element的元素。(15pt)
###       方法2.請利用教學中的numpy 套件中function將Element直接reshape成A(15pt)

## by method 1:  
## Output:


```python
num=0
Element=np.array([1,0,0,0,0,0,3,0,0,0,0,0,5,0,0])
for i in range (0,row): 
    for j in range (0,column):
        zero_A[i,j]=Element[num]
        num+=1
A=zero_A
print("Element =:",Element)
print("matrix A by method 1:\n",A)
print("A's shape is :",A.shape)
```

## Algorithm of diagonal system by for 迴圈:   
    given A, b     
    for i = 1 . . . n    
        x_i = b_i / a_i,i    
    end  
## step 1:given A,b:


```python
v =np.array( [1, 3, 5])
D = np.diag(v)
b =np.array([[-1,6,-15]]).T
x=np.zeros((1,3),dtype=int).T
print(b,"\n",x)
v=np.diag(D)
v=v.reshape((1,3))
v
```

## step 2_v1:by for 迴圈  (20pt)
## Output: x= [-1  2 -3]   
Note:The type of x'output must be int!

```python
m,n=D.shape
for i in range (0,n):
    x[i]=b[i]/D[i,i]
print('x=',x.astype(int))
```
